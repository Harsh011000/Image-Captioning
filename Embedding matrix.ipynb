```python
import numpy as np
import pickle
```

```python
def load_from_disk(path):

    with open(path, "rb") as f:
        loaded_list = pickle.load(f)

    return loaded_list
```

```python
word_to_vec=load_from_disk('./Vocabulary/word_to_vec_dict.pkl')
```

```python
word_to_index=load_from_disk('./Vocabulary/new_words_to_index_dict.pkl')
print('loaded')
```

```python
len(word_to_index)
```

```python
index_to_word=load_from_disk('./Vocabulary/new_index_to_words_dict.pkl')
```

```python
len(index_to_word)
```

```python
embedding_matrix=np.zeros((7828,300))
```

```python
embedding_matrix.shape
```

```python
for x in word_to_index:
    i=word_to_index[x]
    if i==0 or i==1 or i==2 or i==3:
        continue
    emb=word_to_vec[x]
    embedding_matrix[i,:]=emb

print('completed')
```
